```python
# creating a new dataframe without the error from previous df 
clean_df= df.dropna()
# verifying the last couple of rows have no errors
clean_df.tail()
```

```python
# to access a particular column from a df, I can use square braacket notation like so
clean_df["Starting Median Salary"]
```

```python
# to find the highest salary, simply chain the method .max()
clean_df["Starting Median Salary"].max()
```

```python
# we need to know the row number or index so that we can look up the name of the major. 
# the .idxmax() method will give us index for the row with the largest value.
clean_df["Starting Median Salary"].idxmax()
```

```python
# To see the name of the major that corresponds to that particular row, we can use the .loc (location) property.
clean_df["Undergraduate Major"].loc[43]
```

```python
# If you don't specify a particular column you can use the .loc property to retrieve an entire row:
clean_df.loc[43]
```

```python
# calculate the difference between the earnings of the 10th and 90th percentile
spread_col = clean_df["Mid-Career 90th Percentile Salary"].subtract(clean_df["Mid-Career 10th Percentile Salary"])
spread_col
```

```python
# can add the spread column to our existing df with the  folloowing code
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
# to see whic degrees have the smalles spread, use .sort_values()
# and since I'm only interested in seeing the name of the degree we can pass a 
# list of these two column names to look at the .head() of these two columns exclusively.
low_risk = clean_df.sort_values("Spread")
low_risk[["Undergraduate Major", "Spread"]].head()
```
